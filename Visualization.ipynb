```python
import geopandas as gpd
import fiona
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
import matplotlib.gridspec as gridspec
```

```python
warnings.filterwarnings('ignore', 'GeoSeries.notna', UserWarning)
```

```python
huc_4 = gpd.read_file("Source Data/WBD_National_GDB.gdb", driver='FileGDB', layer='WBDHU4')
chesapeake = huc_4[huc_4['HUC4'].isin(['0205', '0206', '0207', '0208'])].unary_union
```

```python
water_geo = gpd.GeoDataFrame(pd.merge(
    left = pd.read_csv("Source Data/Water_FINAL.csv", dtype = 'object'),
    right = gpd.read_file("Source Data/WBD_National_GDB.gdb", driver='FileGDB', layer='WBDHU12'),
    how = 'inner',
    left_on = 'HUC12_',
    right_on = 'HUC12'
))
```

```python
water_geo['YEAR'] = pd.DatetimeIndex(water_geo['Date']).year
```

```python
def plot_coverage(param, year):
    
    fig = plt.figure(constrained_layout=True, figsize = (20, 10))
    gs = fig.add_gridspec(1, 50)
    ax2 = fig.add_subplot(gs[0, :13])
    ax1 = fig.add_subplot(gs[0, 12:])

    ax1.set_axis_off()
    gpd.GeoSeries(chesapeake).plot(ax = ax1, color = 'lightgrey')
    gpd.GeoSeries(chesapeake).boundary.plot(ax = ax1, color = 'k', linewidth = 0.5)


    cbp = gpd.GeoDataFrame(water_geo.loc[(water_geo['Database'] == 'CBP') & (water_geo['ParameterName_CBP'] == param) & (water_geo['YEAR'] == year),'geometry'].unique(), columns = ['geometry'])
    cmc = gpd.GeoDataFrame(water_geo.loc[(water_geo['Database'] == 'CMC') & (water_geo['ParameterName_CBP'] == param) & (water_geo['YEAR'] == year),'geometry'].unique(), columns = ['geometry'])
    cbp.plot(ax = ax1, color = 'lime', alpha = 0.5)
    cmc.plot(ax = ax1, color = 'magenta', alpha = 0.5)
    ratios = [
        cbp.area.sum() / chesapeake.area,
        cmc.area.sum() / chesapeake.area,
        (chesapeake.area - cbp.area.sum() - cmc.area.sum()) / chesapeake.area
    ]

    for s in [cbp, cmc]:
        gpd.GeoSeries(s.unary_union).boundary.plot(ax = ax1, color = 'k', linewidth = 0.5)

    ax2.pie(x = ratios, labels = ['CBP', 'CMC', ''], colors = ['lime', 'magenta', 'lightgrey'], labeldistance = None, wedgeprops = {'linewidth': 0.5, 'edgecolor': 'k'})
    ax2.axis('equal')
    ax2.legend(loc = 'upper center', ncol = 2, fontsize = 'x-large', edgecolor = 'k', bbox_to_anchor=(0.5, 0.9))
    fig.suptitle(f"Watershed Coverage of {param.title()} Data, {year}", fontsize = 20, x= 0.62, y= 1.05)
    fig.savefig(f"Images/Coverage{param}{year}.png", transparent = True, bbox_inches = 'tight')
    plt.close()
```

```python
for param in water_geo['ParameterName_CBP'].unique():
    for year in water_geo['YEAR'].unique():
        try:
            plot_coverage(param, year)
        except:
            plt.close()
            pass
```
